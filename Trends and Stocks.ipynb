```python
from pytrends import dailydata
df_p = dailydata.get_daily_data("buy tesla", 2020, 11, 2021, 11)
df_p2 = dailydata.get_daily_data("elon musk", 2020, 11, 2021, 11)
df_p3 = dailydata.get_daily_data("model 3", 2020, 11, 2021, 11)
df_p4 = dailydata.get_daily_data("model y", 2020, 11, 2021, 11)
df_p5 = dailydata.get_daily_data("tsla stock", 2020, 11, 2021, 11)
df_ne = dailydata.get_daily_data("tesla earnings", 2020, 11, 2021, 11)
df_ne2 = dailydata.get_daily_data("car market", 2020, 11, 2021, 11)
df_ne3 = dailydata.get_daily_data("electric vehicles", 2020, 11, 2021, 11)
df_ne4 = dailydata.get_daily_data("environment", 2020, 11, 2021, 11)
df_n = dailydata.get_daily_data("elon musk fraud", 2020, 11, 2021, 11)
df_n1 = dailydata.get_daily_data("whistleblower", 2020, 11, 2021, 11)
df_n2 = dailydata.get_daily_data("union", 2020, 11, 2021, 11)
df_n3 = dailydata.get_daily_data("tesla debt", 2020, 11, 2021, 11)
df_n4 = dailydata.get_daily_data("subsidy", 2020, 11, 2021, 11)
```
